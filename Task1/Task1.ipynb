```python
import pandas as pd

# Load the dataset
df = pd.read_csv("student-mat.csv", sep=';')

```

```python
df
```

```python
# Display the first few rows
print(df.head())
```

```python
#Data Exploration
#Check for Missing Values:
print(df.isnull().sum())
```

```python
#Display Column Data Types:
print(df.dtypes)
```

```python
#Understand the Dataset's Size:
print(df.shape)
```

```python
 #Data Cleaning
#Handle Missing Values: 
df.dropna(inplace=True)
```

```python
#Remove Duplicate Entries:
df.drop_duplicates()
```

```python
#Data Analysis Questions
#Average Score in Math (G3):
average_score = df['G3'].mean()
print(f'Average score in math (G3): {average_score}')
```

```python
#Students Scoring Above 15 in G3:
above_15_count = df[df['G3'] > 15].shape[0]
print(f'Number of students scoring above 15 in G3: {above_15_count}')
```

```python
#Correlation Between Study Time and G3:
correlation = df['studytime'].corr(df['G3'])
print(f'Correlation between study time and final grade (G3): {correlation}')
```

```python
#Average Final Grade by Gender:
average_by_gender = df.groupby('sex')['G3'].mean()
print(average_by_gender)
```

```python
#Data Visualization
#Histogram of Final Grades (G3):

import pandas as pd
import matplotlib.pyplot as plt

plt.hist(df['G3'], bins=50, color='blue', alpha=0.7)

plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Grades (G3)')
plt.ylabel('Number of Students')

plt.show()
```

```python
#Scatter Plot Between Study Time and G3:
plt.scatter(df['studytime'], df['G3'], alpha=0.5)
plt.title('Study Time vs Final Grade (G3)')
plt.xlabel('Study Time (hours)')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
#Bar Chart Comparing Average Scores of Male and Female Students:
average_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Grade (G3)')
plt.show()
```
